# Regressão Linear com PyTorch

## Objetivo

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se
- o cálculo do gradiente automático do PyTorch, porém
- a função de perda e
- atualização dos parâmetros é feita de forma explícita.
- A rede é criada com uma camada nn.Linear()

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinear_Otimizacao.png)

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
#Inicialização usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight
```

### Testando o predict da rede

```python
xx = torch.ones((5,2))
xx
```

```python
model(Variable(xx))
```

## Treinamento

```python
num_epochs = 100
learning_rate = 0.5
for epoch in range(num_epochs):
    #inputs = Variable(x_train_bias)
    #target = Variable(y_train)
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model(inputs)

    #loss cálculo da função de perda
    loss = torch.mean((y_pred - target)**2)

    # backward propagation - cálculo gradient
    loss.backward()

    # gradiente descendente
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # zeragem parâmetros próximo cálculo do gradiente
    model.weight.grad.data.zero_()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

## Avaliação

```python
y_pred = model(Variable(x_train_bias))
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
```

### Acessando os parâmetros

```python
w_dic = model.state_dict()
w_dic
```

### Parâmetros ótimos, solução analítica

```python
x_bias = x_train_bias
y = y_train
x_bias_t = torch.t(x_bias)
w_opt = (torch.inverse(x_bias_t.mm(x_bias)).mm(x_bias_t)).mm(y)
print(torch.t(w_opt))
```

# Aprendizados

Exercício 1 — Por que loss.data?

```python
# loss é um tensor COM grafo computacional (requires_grad=True)
loss = torch.mean((y_pred - target)**2)

# Sem .data → imprime:
# tensor(0.021170, grad_fn=<MeanBackward0>)

# Com .data → extrai só o valor numérico:
print(loss.data)    # tensor(0.021170)
print(loss.item())  # 0.021170  ← forma mais moderna
```

No final da varredura do laço de treinamento, a perda é impressa utilizando `loss.data` porque o tensor `loss` carrega o grafo computacional do PyTorch, que é necessário para o cálculo dos gradientes no processo de `backward()`.

O atributo `.data` extrai apenas o valor numérico do tensor, sem os metadados relacionados ao autograd, permitindo que o valor da perda seja exibido de forma mais simples e com menor uso de memória.

Entretanto, não se deve utilizar `loss.data.backward()`, pois isso rompe o funcionamento do autograd e impede o cálculo correto dos gradientes.

Exercício 2 — MSE com pesos treinados

```python
# Após o laço de treinamento:
y_pred_final = model(x_train_bias)
loss_final = torch.mean((y_pred_final - y_train)**2)
print(f'MSE com pesos treinados: {loss_final.data:.6f}')
```

Ao calcular e imprimir o valor da perda (MSE) utilizando o predict da rede com os pesos já treinados, obteve-se um MSE final de 0.021170. Esse valor é praticamente idêntico ao da solução analítica (0.021169), o que confirma que o treinamento da rede convergiu corretamente e que os pesos aprendidos estão muito próximos da solução ótima.

Exercício 3 — Listas w0_list e w1_list + gráfico

```python
torch.manual_seed(1234)
model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
model.weight.data = torch.zeros(1, 2)
torch.nn.init.uniform_(model.weight.data, -0.1, 0.1)

w0_list = []
w1_list = []

for epoch in range(num_epochs):
    y_pred = model(x_train_bias)
    loss = torch.mean((y_pred - y_train)**2)
    loss.backward()
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data
    model.weight.grad.data.zero_()

    # Guardar após a atualização
    w0_list.append(model.weight.data[0][0].item())
    w1_list.append(model.weight.data[0][1].item())

# Solução analítica para marcar no gráfico
x_bias = x_train_bias
w_opt = (torch.inverse(x_bias.t().mm(x_bias)).mm(x_bias.t())).mm(y_train)
w0_opt = w_opt[0].item()
w1_opt = w_opt[1].item()


plt.scatter(w0_list, w1_list)
w0_old = None
for (w0, w1) in zip(w0_list, w1_list):
    if w0_old:
        plt.arrow(w0_old, w1_old, w0-w0_old, w1-w1_old,
                  head_length=0.01, head_width=0.01, shape='full',
                  length_includes_head=True)
    w0_old, w1_old = w0, w1


plt.scatter([w0_opt], [w1_opt], color='red', s=200, marker='*', label='Ótimo analítico')
plt.xlabel('w0 (bias)'); plt.ylabel('w1 (coef. angular)')
plt.title('Evolução dos parâmetros durante o treinamento')
plt.legend(); plt.grid(True); plt.show()
```

Ao criar as listas w0_list e w1_list, foi possível armazenar a evolução dos parâmetros durante cada iteração do treinamento e visualizar sua trajetória em um gráfico do tipo scatter plot.

O gráfico mostra que os pesos partem de valores aleatórios próximos de zero e caminham gradualmente em direção ao valor ótimo obtido pela solução analítica, representado pela estrela vermelha. Observa-se que os passos são maiores no início do treinamento, devido ao gradiente mais intenso, e tornam-se menores conforme o modelo se aproxima da solução ótima. Esse comportamento é característico do algoritmo de gradiente descendente, indicando que o modelo convergiu corretamente ao longo do treinamento.
