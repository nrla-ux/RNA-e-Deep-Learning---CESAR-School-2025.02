# Regressão Linear com PyTorch

## Objetivos

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se:
- a função de perda MSE do PyTorch,
- treinamento dos parâmetros via gradiente descendente usando o otimizador.
- A rede é criada com uma camada nn.Linear()

## Importação dos pacotes

````python
%matplotlib inline
import torch
from torch import nn, optim
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

torch.manual_seed(1234)
````

## Leitura dos dados

````python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
````

### Normalização dos dados

````python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
````

````python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
````

````python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
````

## Criação do modelo da rede

````python
model = torch.nn.Linear(2, 1, bias=False)
````

### Verificando a inicialização dos parâmetros

````python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight.data
````

### Testando o predict da rede

````python
model(Variable(torch.ones((5,2))))
````

## Treinamento

### Definindo função de perda e otimizador

````python
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.3)
````

### Laço de treinamento

````python
num_epochs = 120
w0_list = []
w1_list = []
for epoch in range(num_epochs):
    inputs = x_train_bias
    target = y_train

    # forward - predict
    out = model(inputs)

    w0_list.append(model.weight.data[0][0].item())
    w1_list.append(model.weight.data[0][1].item())

    #loss cálculo da função de perda
    loss = criterion(out, target)

    # backward e otimizador
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
````

## Avaliação

````python
y_pred = model(x_train_bias)
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
````

# Aprendizados

Exercício 1 — MSE com criterion

````python
# Após o laço de treinamento:
y_pred = model(Variable(x_train_bias))
loss_final = criterion(y_pred, Variable(y_train))
print(loss_final)
````

Após o treinamento da rede, o valor da função de custo (MSE) foi calculado utilizando a função criterion, a mesma empregada durante o laço de treinamento. O resultado obtido foi um MSE final de 0.021181, valor praticamente idêntico ao da solução analítica (0.021169).

Além disso, utilizar o criterion na etapa de avaliação garante consistência entre o treinamento e a validação do modelo, pois a mesma função de erro é aplicada em ambos os processos.

Exercício 2 — Scatter plot

````python
plt.scatter(w0_list, w1_list)
w0_old = None
for (w0, w1) in zip(w0_list, w1_list):
    if w0_old:
        plt.arrow(w0_old, w1_old, w0-w0_old, w1-w1_old,
                  head_length=0.01, head_width=0.01,
                  shape='full', length_includes_head=True)
    w0_old, w1_old = w0, w1

# Marcar o ótimo analítico
x_bias = x_train_bias
w_opt = (torch.inverse(x_bias.t().mm(x_bias)).mm(x_bias.t())).mm(y_train)
w0_opt = w_opt[0].item()
w1_opt = w_opt[1].item()

plt.scatter([w0_opt], [w1_opt], color='red', marker='*', s=200, label='Ótimo analítico')
plt.xlabel('w0 (bias)'); plt.ylabel('w1 (coef. angular)')
plt.title('Evolução dos parâmetros durante o treinamento')
plt.legend(); plt.grid(True); plt.show()
````

O gráfico scatter plot da evolução dos parâmetros durante o treinamento mostra que, mesmo utilizando o optimizer.step() para realizar as atualizações automáticas dos pesos, os parâmetros percorrem o mesmo caminho de convergência em direção ao ótimo analítico.

Isso confirma que o algoritmo SGD (Stochastic Gradient Descent) implementado automaticamente pelo otimizador é matematicamente equivalente à atualização manual dos pesos realizada no exercício anterior, produzindo a mesma tendência de convergência para a solução ótima.
